```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
#loading the dataset to a pandas dataframe
credit_card_data=pd.read_csv("/content/creditcard[1].csv")
credit_card_data
```

methods to solve the missing values

imputation

dropping(for large data)


```python
new_credit_card_data=credit_card_data.dropna(how='any')
new_credit_card_data
```

```python
new_credit_card_data.isnull().sum()
```

```python
new_credit_card_data.shape
```

```python
#distribution of legit transaction & fraudulent tansaction
new_credit_card_data['Class'].value_counts()
```

this dataset is highly unbalanced

0-->normal transaction

1-->fraudulent transaction

```python
# separating data for analysis
legit=new_credit_card_data[new_credit_card_data.Class==0]
fraud=new_credit_card_data[new_credit_card_data.Class==1]
print(legit.shape)
print(fraud.shape)
```

```python
# statistical measures of the data
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
# compare the values for both transactions
new_credit_card_data.groupby('Class').mean()
```

Under Sampling

Build a sample dataset containing similar distribution of normal transactions and fraudulent transaction

number of fraudulent transactions -->108

```python
legit_sample=legit.sample(n=108)
```

Concatenating two dataframes

```python
new_dataset=pd.concat([legit_sample,fraud],axis=0)
new_dataset
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

Splitting the data into features & targets

```python
x=new_dataset.drop(columns='Class',axis=1)
y=new_dataset['Class']
```

```python
print(x)
```

```python
print(y)
```

Split the data into Training data & Testing data

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,stratify=y,random_state=2)
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

Logistic Regression

```python
from sklearn.preprocessing import StandardScaler
model=LogisticRegression(random_state=0,max_iter=1000)
#training the logistic regression model with training data
model.fit(x_train,y_train)
ypred=model.predict(x_train)
```

Model Evaluation

Accuracy Score

```python
# accuracy on training data
x_train_prediction=model.predict(x_train)
training_data_accuracy=accuracy_score(x_train_prediction,y_train)
```

```python
print("Accuracy on training data :",training_data_accuracy)
```

```python
# accuracy on test data
x_test_prediction=model.predict(x_test)
test_data_accuracy=accuracy_score(x_test_prediction,y_test)
```

```python
print("Accuracy score on test data:",test_data_accuracy)
```
